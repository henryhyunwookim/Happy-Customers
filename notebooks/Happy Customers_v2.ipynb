```python
import pandas as pd
from pycaret.classification import *

from sklearn.model_selection import train_test_split
```

```python
random_state = 1
test_size = 0.05

survey_df = pd.read_csv("./data/ACME-HappinessSurvey2020.csv")
# X = survey_df.drop("Y", axis=1)
# y = survey_df["Y"]
# X_train, X_test, y_train, y_test = \
#             train_test_split(X, y, test_size=test_size, random_state=random_state)
```

```python
train_df = survey_df.sample(frac=0.95, random_state=random_state)
test_df = survey_df.drop(train_df.index)
s = setup(data=train_df, target='Y', session_id=1)
```

```python
best_model = compare_models()
print(best_model)
```

```python
gbc = create_model('gbc')
```

```python
tuned_gbc = tune_model(gbc)
```

```python
tuned_gbc
```

```python
evaluate_model(tuned_gbc)
```

```python
predict_model(tuned_gbc)
```

```python
final_gbc = finalize_model(tuned_gbc)
```

```python
predict_model(final_gbc)
```

```python
test_pred = predict_model(final_gbc, test_df)

from pycaret.utils import check_metric
check_metric(test_pred['Y'], test_pred['Label'], metric = 'Accuracy')
```
